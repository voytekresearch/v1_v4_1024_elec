### Load data

```python
# load LFP spectral params results by bands
results = pd.read_csv(fr"{EXTERNAL_PATH}\data\results\lfp_spectral_params_bands.csv", index_col=0)
```

### Compute stimulus-evoked change in periodic params

```python
# split into two dataframes (enconding and baseline)
results_pre = results.loc[results["epoch"]=="pre"]
results_post = results.loc[results["epoch"]=="post"]
```

```python
# merge two dataframes with suffixes for baseline and encoding
df_merge = results_post.merge(results_pre, on=['session','channel','chan_idx','array'], suffixes=['_post', '_pre'])
df_merge = df_merge.drop(columns=['epoch_pre','epoch_post'])

# compute difference
varname = ['offset', 'knee', 'exponent', 'r_squared', 'alpha_cf', 'alpha_bw', 'alpha_pw', 'beta_cf', 'beta_bw', 'beta_pw', 'gamma_cf', 'gamma_bw', 'gamma_pw']
for var in varname : 
    df_merge[f'{var}_diff'] = df_merge[f'{var}_post'] - df_merge[f'{var}_pre']

df_merge.head()
```

```python
# save results (overwrite spectral_params_bands.csv)
df_merge.to_csv(f'{EXTERNAL_PATH}/data/results/lfp_spectral_merged_bands.csv')
```

### Plot


```python
bands = ['alpha', 'beta', 'gamma']

# plot configuration
pos = [1, 2, 3]
label = ['cf', 'bw', 'pw']

for b in bands:
    # create dataframe for violin plot
    band_df = pd.DataFrame()
    band_df['cf'] = df_merge[f'{b}_cf_diff'].dropna()
    band_df['pw'] = df_merge[f'{b}_pw_diff'].dropna()
    band_df['bw'] = df_merge[f'{b}_bw_diff'].dropna()

    # plot
    fig, ax = plt.subplots()
    ax.violinplot(band_df)
    ax.set_title(f"\u0394 {b}")
    ax.set_xticks(pos)
    ax.set_xticklabels(label)
    ax.axhline(y=0, color='k', linestyle='--')
    plt.show()

    # save figure
    fig.savefig(fr"G:\Shared drives\v1_v4_1024\figures\periodic\{b}_diff.png")
```

```python
# plot violin of both pre- and post- stimulus spectral params

for band in ['alpha', 'beta', 'gamma']:
    for param, label in zip(['cf', 'pw', 'bw'], ['center frequency', 'relative power', 'bandwidth']):
        # plot
        fig, ax = plt.subplots()
        sns.violinplot(data=results, x="session", y=f"{band}_{param}", hue="epoch", split=True, inner='quarts', ax=ax)
        ax.set_title(f"{band} - {label}")
        ax.set_ylabel(label)

        # save figure
        fig.savefig(fr"G:\Shared drives\v1_v4_1024\figures\periodic\{band}_{param}.png")
```
